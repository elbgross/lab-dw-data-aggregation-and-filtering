1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
import pandas as pd
import numpy as np

df= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
print(df)
# Convert 'Total Claim Amount' to numeric, forcing errors to NaN, then round to 2 decimal places
df['Total Claim Amount']= pd.to_numeric(df['Total Claim Amount'], errors='coerce')
df['Total Claim Amount']= df['Total Claim Amount'].astype(float).round(2)
print(df['Response'].unique())
resp_yes= df['Response'].astype(str).str.contains('Yes')
df_claim_yes=df[resp_yes & (df['Total Claim Amount'] < 1000)]
df_claim_yes

```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
df['Monthly Premium Auto']= pd.to_numeric(df['Monthly Premium Auto'], errors='coerce')
df['Customer Lifetime Value']= pd.to_numeric(df['Customer Lifetime Value'], errors='coerce')
resp_yes= df['Response'].astype(str).str.contains('Yes')
print(df['Policy Type'].unique())
print(df['Gender'].unique())
df_mean_monthly= df.pivot_table(index='Gender', columns='Policy Type', values='Monthly Premium Auto', aggfunc=np.mean).round(2)
df_mean_monthly
df_mean_lifetime= df.pivot_table(index='Gender', columns='Policy Type', values='Customer Lifetime Value', aggfunc=np.mean).round(2)
df_mean_lifetime

```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
print(df['State'].unique())
customers_per_state= df.groupby('State').size().reset_index(name='Number of Customers')
print(customers_per_state)
customers_over_500= customers_per_state[customers_per_state['Number of Customers'] > 500]
print(customers_over_500)
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
df_lifetime_stats= df.pivot_table(index='Gender', columns='Education', values='Customer Lifetime Value', aggfunc=[np.median, np.min, np.max]).round(2)
df_lifetime_stats
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
df['month']= df['Effective To Date'].str.split('/').str[0]

avg_policies=df.pivot_table(index='State',columns='month',values='Number of Policies', aggfunc='count')
avg_policies
```
